# Data Collection

```python
# Download Original Dataset from WHO
!wget https://covid19.who.int/WHO-COVID-19-global-data.csv

# Read Data
import pandas as pd
df = pd.read_csv("WHO-COVID-19-global-data.csv")

# Display Top Data in Dataset
df.head()
```

# Exploratory Data Analysis (EDA)

**Check Null Value per Columns**

```python
# Check Null Value per Columns
df.isnull().sum()
```

**Delete Rows if Have Null Value**

```python
# Delete rows if have null value
df = df.dropna()

# Check dataframe again
df.isnull().sum()
```

**Delete Unused Columns**

```python
# Delete Unused Columns in Dataframe
df = df.drop(['Country_code','WHO_region', 'Cumulative_cases', 'Cumulative_deaths'], axis=1)
```

```python
# Display Dataframe
df.head()
```

**Get Data Only Indonesia**

```python
# Get data with condition
df1 = df[df['Country'] == "Indonesia"]
# Display new dataframe
df1.head()
```

```python
# Reset index dataframe
df1.reset_index(drop=True, inplace=True)
# Display dataframe again
df1.head()
```

**Create Visualization for New Cases**

```python
# Import Libraries
import plotly.express as px
# Setup plot
fig = px.bar(df1, x='Date_reported', y='New_cases')
# DIsplay plot
fig.show()
```

**Create Visualization for New Deaths**

```python
# Setup plot
fig = px.bar(df1, x='Date_reported', y='New_deaths')
# DIsplay plot
fig.show()
```

**Create Visualization for Cases vs Deaths**

```python
# Calculate data value
cases = df1["New_cases"].sum()
deaths = df1["New_deaths"].sum()

# Customize plot
labels = ["Cases", "Deaths"]
values = [cases, deaths]
fig = px.pie(df1, values=values, names=labels, title='Total Cases vs Total Deaths', hole=0.5)

# Display plot
fig.show()
```

# Prediction Death

```python
# Import Libraries for Forecast
from autots import AutoTS
# Create Model
model = AutoTS(forecast_length=30, frequency='infer', ensemble='simple') #30 => 30 days prediction
# Fit Model
model = model.fit(df1, date_col="Date_reported", value_col='New_deaths', id_col=None)
# Start Predict
prediction = model.predict()
forecast = prediction.forecast
# Display forecast
print(forecast)
```
